```python
"""
字典和集合基础
字典由 键 和 值 组成
字典有序
字典的性能更优，特别是查 添 删
集合无序，不可重复
"""
# 常见创建方法
d1 = {'name': 'Tom', 'age': 20}
d2 = dict({'name': 'Tom', 'age': 20})
d3 = dict([('name', 'Tom'), ('age', 20)])
d4 = dict(name='Tom', age=20)
print(d1)
print(d2)
print(d3)
print(d4)
s1 = {1,2,3}
s2 = set([1,2,3])
print(s1)
print(s2)
```

```python
# 访问
d1['name']
d1.get('1','NULL')  # 如果不存在，返回默认值
# 集合不支持索引操作
```
