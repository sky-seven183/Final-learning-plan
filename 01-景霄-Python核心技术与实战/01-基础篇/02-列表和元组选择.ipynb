```python
"""
两种基本数据结构：列表和元组
基础：可以放置任意数据类型的有序集合
区别：
列表是动态的，可以随时添加和删除元素
元组是静态的，长度大小固定，无法增加删减或者改变
"""
list1 = [1,2,3,4]
list1.append(5)
print(list1)
tuple1 = (1,2,3,4)
new_tuple = tuple1 + (5,)
print(new_tuple)
print(list1[3])
```

```python
"""
基本操作和注意事项
"""
# 负数索引
print(list1[-1],tuple1[-1])
# 支持切片
print(list1[::-1],tuple1[:3:2])
# 可以随意嵌套
l = [[1,3,5,7,9],[2,4,6,8,10]]
t = ((1,2,3),(4,5,6))
print(t[0],l[0][0],t,l)
# 支持强制转换
print(list(t))
# 内置函数
list1.count(1)  # 统计元素出现次数
list1.index(1)  # 返回元素索引,第一次出现
list1.reverse()  # 原地反转列表，元组没有
list1
list1.sort()   # 原地排序列表，元组没有
list1
tuple1.count(4)
tuple1.index(4)
#  reserved() 和 sorted() 同样表示对列表/元组的倒转和排序,但有返回值
```

```python
"""
存储方式差异:
列表是动态的，1需要存储指针,2额外分配存储空间,防止存储空间不够
    保证了操作的高效性,额外空间避免了频繁的内存申请和释放
性能的差异:
    元组的量级更轻,性能速度略微犹豫列表
    Python在后台对静态数据做资源缓存,垃圾回收机制,回收不在使用的变量,返回给操作系统
    但对于静态变量,不被使用且空间不大时,Python会暂时缓存,下一次创建相同的元组时,直接分配之前的空间,大大加快程序的运行速度

    初始化元组快5倍,索引差距小,忽略不计

"""
```

```python
"""
使用场景
存储的数据和数量不变,元组合适
存储的数据或数量可变,列表更适合
"""
```
